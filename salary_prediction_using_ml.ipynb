```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
x=df["Experience Years"]
y=df["Salary"]
plt.figure(figsize=(10,3))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
((32,1), (8,1), (32), (8))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train.values.reshape(-1,1),y_train)
```

```python
lr.score(x_train.values.reshape(-1,1),y_train)*100
```

```python
year=float(input("Enter your year of experience:"))
lr.predict([[8]])
```

```python
x=df["Experience Years"]
y=df["Salary"]
plt.figure(figsize=(10,3))
plt.scatter(x,y)
plt.plot(x,lr.predict(x.values.reshape(-1,1)),color="red")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.show()
```
